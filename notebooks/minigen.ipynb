```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import sys
sys.path.append(r"../src/")
from party.histogram import *
```

```python
eta_dist = np.loadtxt("../minigen/outputs/ee_y_mumu_eta_dist.txt")
eta_dist
```

```python
h_eta = Hist1D(40, -5., 5.)
h_eta.set_error_type(BinErrorType.Normal)
h_eta.fill_array(eta_dist)
```

```python
plt.figure()
h_eta.plot(label=r"$ee\to\gamma\to\mu\mu$", color="blue", density=False)
plt.xlabel(r"$\eta(\mu)$")
plt.ylabel("Events")
plt.legend()
plt.show()
```
