```python
import sys
sys.path.append(r"../src/")
from party.histogram import *
```

```python
h1 = Hist1D(2, 0., 2.)
h1.set_error_type(BinErrorType.Normal)

h2 = Hist1D(2, 0., 2.)
h2.set_error_type(BinErrorType.Normal)
```

```python
h1.fill_array([0, 0, 0, 1, 1])
h2.fill_array([0, 0, 1, 1, 1])
h3 = h1 * h2
h4 = h1 / h2
h5 = h1 * h2 / h2
```

```python
plt.figure()
h1.plot(label="Histogram 1", color="black")
h2.plot(label="Histogram 2", color="red")
plt.xlim([0, 2])
plt.xlabel("Variable Name")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
plt.figure()
h3.plot(label="Hist 1 x Hist 2", color="black")
h4.plot(label="Hist 1 / Hist 2", color="red")
plt.xlim([0, 2])
plt.xlabel("Variable Name")
plt.ylabel("Count")
plt.legend()
plt.show()
```
