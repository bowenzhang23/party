```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ecm, xs, xs_err = \
    np.loadtxt("../minigen/outputs/uu_zy_mumu_ecm.txt"), \
    np.loadtxt("../minigen/outputs/uu_zy_mumu_xs.txt"), \
    np.loadtxt("../minigen/outputs/uu_zy_mumu_xs_err.txt")
```

```python
plt.figure()
plt.errorbar(
    ecm,
    xs,
    yerr=xs_err,
    fmt="none",
    ecolor="k",
    label=r"$uu\to Z/\gamma\to\mu\mu$",
    color="blue",
)
plt.xlabel(r"$\sqrt{s}$ [GeV]")
plt.ylabel("Cross section [pb]")
plt.legend()
plt.yscale("log")
plt.show()
```
