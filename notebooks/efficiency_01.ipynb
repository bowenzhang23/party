```python
import sys
sys.path.append(r"../src/")
import matplotlib.pyplot as plt
from party.histogram import Hist1D, BinErrorType
from party.efficiency import Efficiency1D
```

```python
h1 = Hist1D(2, 0., 2.)
h1.set_error_type(BinErrorType.Normal)

h2 = Hist1D(2, 0., 2.)
h2.set_error_type(BinErrorType.Normal)
```

```python
h1.fill_array([0, 1, 1])
h2.fill_array([0, 0, 1, 1, 1])
eff = Efficiency1D(h1, h2)
```

```python
plt.figure()
eff.plot(label="Efficiency", color="black")
plt.ylim([0, 1])
plt.xlabel("Variable Name")
plt.ylabel("Efficiency")
plt.legend()
plt.show()
```
